### ( 8 )
##### For each lender, compute the overall amount of money lent. For each loan that has more than one lender, you must assume that all lenders contributed the same amount.

Numero di lenders per ogni prestito nella tabella di destra del merge, a sinistra la tabella non aggregata. In questo modo, posso fare un altra left join e recuperare per ogni coppia presitito-lender l'importo totle del prestito e poi e' sufficiente dividere

```python
number_loansLenders = (loans_lenders_stacked.merge(loans_lenders_stacked.groupby('loan_id').agg('count').reset_index(),
                                     how = 'left',
                                     on = ['loan_id']))
number_loansLenders.columns = ['loan_id', 'lenders', 'count']
number_loansLenders.head()
```

```python
#join
merged_loans = number_loansLenders.merge(loans[['loan_id','loan_amount']],
                                         how = 'left',
                                         on = 'loan_id')
merged_loans.head()
```

Ottengo l'importo prestato da ogni lender, e poi aggrego per ottenere il totale

```python
merged_loans['each_money'] = merged_loans['loan_amount'] / merged_loans['count']
lenders_totalAmounts = pd.DataFrame(merged_loans.groupby('lenders')['each_money'].sum())
lenders_totalAmounts.head(20)
```

Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st, 2016, planned expiration time January 30th 2018, and amount 5000USD has an amount of 5000USD * 31 / (31+365+30) = 363.85 for 2016, 5000USD * 365 / (31+365+30) = 4284.04 for 2017, and 5000USD * 30 / (31+365+30) = 352.11 for 2018.

Funzione che restituisce una lista di due liste, la prima contenente gli anni di riferimento e la seconda contenente l'importo di competenza per l'anno associato alla corrispondente posizione dell'altra lista.

Il modo piu 'elegante' per fare questo tipo di operazione sarebbe stato quello di creare una lista di dataframe e utilizzare pd.concat() che pero' risulta molto meno performante del dizionario (almeno un'ora contro max 5 minuti).

```python
def getWeights(first, second, amount):
    
    #per evitare problemi dovuti agli orari, si fissa tutto a mezzanotte
    first = first.replace(hour = 0, second = 0, minute = 0)
    second = second.replace(hour = 0, second = 0, minute = 0)
    
    #se la data di scadenza è antecedente alla data di esborso, il calcolo non viene fatto e si passa al prestito successivo
    if second < first:
        return [[],[]]
        #alternativamente si poteva scambiare data esborso con data scadenza ipotizzando che in fase di valorizzazione siano state
        #erroneamente scambiate. per farlo basta utilizzare la seguente riga di codice:
        #first, second = second, first
    
    #inizializzazione dell'output e dei pesi
    years_list = []
    amount_list = []
    w = []
    #numero totale di giorni
    gg = (second - first).days + 1
    final_amount = 0
    
    for anno in range(first.year, second.year + 1):
        
        #esborso e scadenza cadono in anni diversi
        if first.year != second.year:
            
            #anno che sta fra i due estremi -> il peso è 365 o 366, in base al fatto che sia o meno bisestile
            if (anno != first.year) & (anno != second.year):
                if anno % 4:
                    weight = 365
                else:
                    weight = 366
            #se sto calcolando l'importo dell'anno della data di esborso, calcolo il numero di giorni che stanno fra quella data e
            #l'ultimo giorno dell'anno (ultimo giorno compreso, quindi e' come fare primo giorno dell'anno successivo)
            elif anno == first.year:
                weight = (first.replace(year = first.year + 1, month = 1, day = 1) - first).days
            #se sto calcolando l'importo dell'anno della data di esborso, calcolo il numero di giorni che stanno fra quella data e
            #il primo giorno dell'anno (primo giorno compreso, quindi e' come fare ultimo giorno dell'anno precedente)
            elif anno == second.year:
                weight = (second - second.replace(year = second.year - 1, month = 12, day = 31)).days
        #esborso e scadenza cadono nello stesso anno    
        else:
            #il peso e' il numero di giorni che stanno fra le due date e cade nell'anno delle due date
            weight = gg
        
        #l'importo dell'anno e' proporzionale al numero di giorni di competenza di quell'anno
        current_amount = amount * weight / gg
        
        #aggiungo l'importo e anno corrente agli output
        years_list = years_list + [anno]
        amount_list = amount_list + [current_amount]
        
    output_data = [years_list, amount_list]
    
    #debug: l'importo totale e' pari all'importo suddiviso nei diversi anni?
    #assert (round(sum(output_data[1])) == amount), 'the amounts do not match each other'
    
    return output_data
```
